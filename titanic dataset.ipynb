```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Load the dataset
data = pd.read_csv("C:/Users/patha/OneDrive/Desktop/future intern/Clean data/cleaned_data_test.csv")
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())
```

```python
# Calculating summary statistics
def calculate_summary_statistics(df, columns):
    summary_stats = {}
    for col in columns:
        summary_stats[col] = {
            "mean": df[col].mean(),
            "median": df[col].median(),
            "mode": df[col].mode()[0],
            "std_dev": df[col].std(),
        }
    return summary_stats

# Specify the columns for which to calculate statistics
columns_to_analyze = ['Age', 'SibSp', 'Parch', 'Fare'] 

# Call the function and get the results
stats = calculate_summary_statistics(data, columns_to_analyze)

# Print the statistics
for col, stat in stats.items():
    print(f"\nSummary Statistics for {col}:")
    print(f"Mean: {stat['mean']}")
    print(f"Median: {stat['median']}")
    print(f"Mode: {stat['mode']}")
    print(f"Standard Deviation: {stat['std_dev']}")
```

```python
# Summary statistics for categorical columns
def categorical_summary(df, columns):
    for col in columns:
        print(f"\nValue counts for {col}:")
        print(df[col].value_counts())
        print(f"Mode for {col}: {df[col].mode()[0]}")
# Specify categorical columns to analyze
categorical_columns = ['Pclass', 'Sex_male', 'Embarked_Q', 'Embarked_S']

# Call the function
categorical_summary(data, categorical_columns)
```

```python
# Correlation matrix for numerical columns
print("\nCorrelation matrix for numerical columns:")
correlation_matrix = data[['Age', 'SibSp', 'Parch', 'Fare']].corr()
print(correlation_matrix)

#visulise the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Histogram for numerical columns
def plot_histograms(df, columns):
    df[columns].hist(bins=20, figsize=(12, 8), edgecolor='black')
    plt.suptitle('Histograms of Numerical Columns', fontsize=16)
    plt.show()
# Call the function for specific columns
plot_histograms(data, ['Age', 'SibSp', 'Parch', 'Fare'])
```

```python
# Boxplot for numerical columns
def plot_boxplots(df, columns):
    df[columns].plot(kind='box', subplots=True, layout=(2, 2), figsize=(12, 8), grid=True)
    plt.suptitle('Boxplots for Numerical Columns', fontsize=16)
    plt.show()
# Call the function for specific columns
plot_boxplots(data, ['Age', 'SibSp', 'Parch', 'Fare'])
```

```python
# Group by and aggregate statistics
grouped_stats = data.groupby('Pclass').agg({
    'Age': ['mean', 'median', 'std'],
    'Fare': ['mean', 'median', 'std'],
    'SibSp': ['sum'],
    'Parch': ['sum']
})
print("\nGrouped summary statistics by Pclass:")
print(grouped_stats)
```

```python
# Check for missing values
print("\nMissing values in each column:")
print(data.isnull().sum())

# Filling missing values with the mean or median
data['Age'].fillna(data['Age'].mean(), inplace=True)
```

```python
# Z-score for detecting outliers in 'Fare'
data['Fare_zscore'] = np.abs(stats.zscore(data['Fare']))

# Print rows with Z-score greater than 3 (potential outliers)
outliers = data[data['Fare_zscore'] > 3]
print("\nOutliers based on Z-score for Fare:")
print(outliers[['PassengerId', 'Fare', 'Fare_zscore']])
```

```python
# Outlier detection function using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))]
    return outliers

# Detect outliers for a specific column (e.g., 'Age')
outliers_age = detect_outliers_iqr(data, 'Age')

# Display the outliers
print(outliers_age)
```

```python
# Z-score normalization of numerical columns
z_score_normalized = data[numerical_columns].apply(stats.zscore)

# Display the first few rows of the normalized data
print(z_score_normalized.head())
```
