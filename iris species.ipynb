```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
iris_data = pd.read_csv("C:/Users/patha/OneDrive/Desktop/future intern/dataset iris species/Iris.csv")

# Preview the dataset
print(iris_data.head())
```

```python
#Plot histograms for each numerical feature
plt.figure(figsize=(16, 10))
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
# Create histograms
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.histplot(iris_data[feature], kde=True, bins=20, color='skyblue')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Bar chart for species distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Species', data=iris_data, palette='Set2')
plt.title('Species Distribution')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# Pairplot to visualize the relationships between features
sns.pairplot(iris_data, hue='Species', diag_kind='kde', palette='Set1')
plt.show()
```

```python
# Boxplot to visualize distribution and outliers for each feature by species
plt.figure(figsize=(16, 10))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='Species', y=feature, data=iris_data, palette='Set3')
    plt.title(f'Boxplot of {feature} by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
# Violin plot to visualize distribution of each feature across species
plt.figure(figsize=(16, 10))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.violinplot(x='Species', y=feature, data=iris_data, palette='muted')
    plt.title(f'Violin Plot of {feature} by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
# Heatmap to show correlation between features
plt.figure(figsize=(8, 6))
corr_matrix = iris_data[features].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# Swarm plot to visualize the distribution of each feature across species
plt.figure(figsize=(16, 10))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.swarmplot(x='Species', y=feature, data=iris_data, palette='Dark2')
    plt.title(f'Swarm Plot of {feature} by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
# Joint plot for SepalLengthCm vs SepalWidthCm
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=iris_data, kind='scatter', hue='Species', palette='Set1')
plt.show()

# Joint plot for PetalLengthCm vs PetalWidthCm
sns.jointplot(x='PetalLengthCm', y='PetalWidthCm', data=iris_data, kind='scatter', hue='Species', palette='Set1')
plt.show()
```

```python
# FacetGrid for Sepal Length across species
g = sns.FacetGrid(iris_data, col="Species", height=4, aspect=1)
g.map(sns.histplot, 'SepalLengthCm', kde=True)
plt.show()

# FacetGrid for Petal Length across species
g = sns.FacetGrid(iris_data, col="Species", height=4, aspect=1)
g.map(sns.histplot, 'PetalLengthCm', kde=True)
plt.show()
```

```python
from pandas.plotting import andrews_curves
plt.figure(figsize=(10, 6))
andrews_curves(iris_data, 'Species', colormap='cool')
plt.title('Andrews Curves')
plt.show()
```
